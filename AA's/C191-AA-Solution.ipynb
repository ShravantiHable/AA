```python
#Addition-Activity-1
#Find the total number of successful Missions as per the company and create a dataframe out of it

#predefine start
import pandas as pd 
import matplotlib .pyplot as plt

dataframe = pd.read_csv("space_Corrected.csv")
df=dataframe.dropna()
df
#predefine end

group_by_success= df.loc[df['Status Mission'] == 'Success']
group_by_success
count_success = group_by_success.groupby('Company Name')['Status Mission'].count().reset_index(name='Status')
count_success

```

```python
#Addition-Activity-2
#Plot a bar graph to show the number of successful missions of the companies

#predefine start
import pandas as pd 
import matplotlib .pyplot as plt

dataframe = pd.read_csv("space_Corrected.csv")
df=dataframe.dropna()
df

group_by_success= df.loc[df['Status Mission'] == 'Success']
group_by_success
count_success = group_by_success.groupby('Company Name')['Status Mission'].count().reset_index(name='Status')
count_success
print(count_success)

#predefine end

plt.xlabel("Company Name")
plt.xticks(rotation='vertical')
plt.ylabel("Number Of Activie Rockets")

label = count_success['Company Name']
value = count_success['Status']

plt.bar(label, value,width=0.4, color=('red','blue','green','pink','yellow')) #bar-grap
```

```python
#Addition-Activity-4
#Which Rockets has the maximum Cost of the mission
#Which Rockets has the minimum Cost of the mission

#predefine start
import pandas as pd 
import matplotlib .pyplot as plt

dataframe = pd.read_csv("space_Corrected.csv")
df=dataframe.dropna()
df

new = df.rename(columns={" Rocket": "Cost_Of_Mission"})
sorted_df = new.sort_values(by=['Cost_Of_Mission'])
sorted_df
#predefine end

maximum = sorted_df.head(1)
print("Maximum : ")
print("Rocket Name : " , maximum['Detail'] , " , Cost - " , maximum['Cost_Of_Mission'])

minimum = sorted_df.tail(1)
print("Minimum : ")
print("Rocket Name : " , minimum['Detail'] , " , Cost - " , minimum['Cost_Of_Mission'])
```
