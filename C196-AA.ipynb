```python
#Additional Activity 4
#predefine code
#import the required packages 
import pandas as pd
import matplotlib.pyplot as plt
#Read the csv file.
df = pd.read_csv("titanic_crew_info.csv")
df
df = df.dropna()
df
new_df = df.loc[df['Survived'] == 1.0]
new_df
groupby_age = new_df.groupby('Age')['Name'].count().reset_index()
groupby_age
#predefine code end


plt.figure(figsize=(7,7))
plt.hist(groupby_age['Name'].head(50), bins=10, label="Age")


plt.title("Survived crew age group ")
plt.ylabel("Survived")
plt.xlabel("Age group")
plt.legend()
plt.show()
```

```python
#Additional Activity 5
#predefine code
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic_crew_info.csv')
df
#predefine code end

survived = df.loc[df['Survived'] == 0.0]
survived

group_position = survived.groupby('Position')['Name'].count().reset_index()
group_position

value = group_position['Name']
label = group_position['Position']

plt.figure(figsize=(15,10))

#predefine code
plt.xlabel("Position")
plt.xticks(rotation='vertical')
plt.ylabel("Deaths")
#predefine code end


plt.bar(label, value,width=0.4, color=('red','blue','green','pink','yellow')) #bar-grap
```
